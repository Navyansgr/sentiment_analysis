```python
# Install required libraries
!pip install textblob

import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Sample dataset (manually entered tweets)
sample_tweets = [
    ["The new policy is amazing!", "Positive"],
    ["I hate how slow the system is", "Negative"],
    ["Itâ€™s just okay, nothing special", "Neutral"],
    ["Great initiative by the government!", "Positive"],
    ["Worst decision ever made", "Negative"],
    ["I love the new tax reforms!", "Positive"],
    ["The service is getting worse every day", "Negative"],
    ["Nothing much has changed, honestly.", "Neutral"]
]

# Convert to DataFrame
df = pd.DataFrame(sample_tweets, columns=["Tweet", "Sentiment"])

# Display sentiment count
print(df["Sentiment"].value_counts())

# Visualization
df["Sentiment"].value_counts().plot(kind="bar", color=["green", "red", "blue"])
plt.title("Sentiment Analysis of Sample Tweets")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```
